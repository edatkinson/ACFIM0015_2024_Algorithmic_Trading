```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st 
from scipy.stats import ttest_rel, wilcoxon, shapiro, pearsonr, binom_test
import pandas as pd
import seaborn as sns
from statistical_tests import get_data, run_statistical_tests
from collections import defaultdict
```

```python

# #Plot box plots for the data files
metrics = ['last_bal', 'n_trades']
phases = ['phase1', 'phase2', 'phase3']
data_trades = pd.DataFrame()
data_bals = pd.DataFrame()
data_max_prof = pd.DataFrame()
for phase in phases:
    trades = get_data('n_trades', phase)
    data = get_data('last_bal', phase)
    data_max = get_data('max_prof', phase)

    data_trades = pd.concat([data_trades, trades[1:]])
    data_bals = pd.concat([data_bals, data[1:]])
    data_max_prof = pd.concat([data_max_prof, data_max[1:]])


#trades:
n_trades = 'n_trades'
pt1_trades = data_trades[f'{n_trades}_values_1']
pt2_trades = data_trades[f'{n_trades}_values_2']

# final bals
final_bals = 'last_bal'
pt1_bal = data_bals[f'{final_bals}_values_1'] # remember this for each scenario
pt2_bal = data_bals[f'{final_bals}_values_2']

max_profs = 'max_prof'
pt1_max = data_max_prof[f'{max_profs}_values_1']
pt2_max = data_max_prof[f'{max_profs}_values_2']

pt1_bal.reset_index(drop=True, inplace=True)
pt2_bal.reset_index(drop=True, inplace=True)
pt1_trades.reset_index(drop=True, inplace=True)
pt2_trades.reset_index(drop=True, inplace=True)

pt1_max.reset_index(drop=True, inplace=True)
pt2_max.reset_index(drop=True, inplace=True)



```

```python
pt1_vals = [np.array(pt1_bal[i][0]) for i in range(0, len(pt1_bal))] #l of l for balances 
pt2_vals = [np.array(pt2_bal[i][0]) for i in range(0, len(pt2_bal))]

pt1_session_trades = [pt1_trades[i][0] for i in range(0, len(pt1_trades))] #l of l for trades in each session for all scenarios
pt2_session_trades = [pt2_trades[i][0] for i in range(0, len(pt2_trades))]

pt1_final_drawdown = [np.array(pt1_max[i][0]) - np.array(pt1_bal[i][0]) for i in range(0,len(pt1_trades))]
pt2_final_drawdown = [np.array(pt2_max[i][0]) - np.array(pt2_bal[i][0]) for i in range(0,len(pt2_trades))]

pt1_scenarios = [pt1_trades[i] for i in range(1, len(pt1_trades))]
data
# pt1_vals = [np.mean(pt1_bal[i][0]) for i in range(1, len(pt1_bal))] #l of l for balances 
# pt2_vals = [np.mean(pt2_bal[i][0]) for i in range(1, len(pt2_bal))]

# pt1_session_trades = [np.mean(pt1_trades[i][0]) for i in range(1, len(pt1_trades))] #l of l for trades
# pt2_session_trades = [np.mean(pt2_trades[i][0]) for i in range(1, len(pt2_trades))]
```

```python
##ppts
ppt1_list = []
ppt2_list = []
for i in range(len(pt1_vals)):
    profits1 = pt1_vals[i] 
    trades1 = np.dot(pt1_session_trades[i], 1/2)
    
    profits2 = pt2_vals[i]
    trades2 = np.dot(pt2_session_trades[i], 1/2)

    ppt1 = np.round(profits1 / trades1)
    ppt2 = np.round(profits2 / trades2)

    ppt1_list.append(ppt1)
    ppt2_list.append(ppt2)


```

```python
##sharpe ratios!
avg_ppt1 = [np.mean(ppt1) for ppt1 in ppt1_list]
avg_ppt2 = [np.mean(ppt2) for ppt2 in ppt2_list]
std_ppt1 = [np.var(ppt1) for ppt1 in ppt1_list]
std_ppt2 = [np.var(ppt2) for ppt2 in ppt2_list]


sharpe_1 = np.array([avg_ppt1[i] / np.sqrt(std_ppt1[i]) for i in range(len(avg_ppt1))] )
sharpe_2 = np.array([avg_ppt2[i] / np.sqrt(std_ppt2[i]) for i in range(len(avg_ppt2))] )

sharpe_1
```

```python
### max_drawdown
pt1_drawdowns = [i.mean() for i in pt1_final_drawdown]
pt2_drawdowns = [i.mean() for i in pt2_final_drawdown]
```

```python
scenarios_ids = data_bals['scenario_id'][0:]
scenarios_ids
```

```python
stats_and_p = np.zeros((2,len(ppt1_list)))
for i in range(len(ppt1_list)-1):
    stat, p_val = run_statistical_tests('ppt', ppt2_list[i], ppt1_list[i], direction='greater')
    stats_and_p[:, i] = np.array([stat, p_val])

stats_and_p[:,0]

n_trials = [500 for _ in range(6)]
for _ in range(19):
    n_trials.append(200)
```

```python
df = pd.DataFrame([scenarios_ids.to_list(),sharpe_1.tolist(), avg_ppt1, std_ppt1, pt1_drawdowns, sharpe_2.tolist(), avg_ppt2, std_ppt2, pt2_drawdowns,stats_and_p[0, :].tolist(), stats_and_p[1,:].tolist(), n_trials]).transpose()
df.columns = ['Scenario','Sharpe_PT1', 'Median_PPT_PT1','Std_PPT_PT1','final_drawdown1', 'Sharpe_PT2' ,'Median_PPT_PT2','Std_PPT_PT2', 'final_drawdown2','W-Stat', 'p-value', 'n_trials']
df
```

```python
# Define PT2 wins: Median_PPT_PT2 > Median_PPT_PT1
pt2_wins = (df['Median_PPT_PT2'] > df['Median_PPT_PT1']).sum()
total_scenarios = len(df)

print(f"PT2 wins in {pt2_wins} out of {total_scenarios} scenarios.")

# Run the binomial test
p_value = binom_test(pt2_wins, total_scenarios, p=0.5, alternative='greater')

print(f"Binomial test p-value: {p_value}")
```
